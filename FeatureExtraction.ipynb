```python
def extract_product_features(description, reviews):
    baseline = """
    I am a grad student in a Generative AI lab course. In this course, we have an 
    assignment to choose some products from Amazon, take their description and  
    reviews as textual data, and eventually create an accurate image of the product using
    generative image AI. I will give you all the original textual content from a 
    product. When I finish and type 'DONE', extract all the information 
    about what the product looks like so that I can create an accurate image later.
    Focus on visual information such as colors, texture, accessories, shape,
    decoration details, etc. We want to generate an image of what the product looks
    like in it's ideal state like in the Amazon-supplied images. 
    """
    completion = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": baseline},
            {"role": "user", "content": description},
            {"role": "user", "content": reviews},
            {"role": "user", "content": "DONE"},
        ]
    )
    print(completion.choices[0].message)
    print()
    
def extract_features_description(description):
    baseline = """
    I am a grad student in a Generative AI lab course. In this course, we have an 
    assignment to choose some products from Amazon, take their description as textual 
    data, and eventually create an accurate image of the product using
    generative image AI. I will give you all the original textual content from a 
    product. When I finish and type 'DONE', extract all the information 
    about what the product looks like so that I can create an accurate image later.
    Focus on visual information such as colors, texture, accessories, shape,
    decoration details, etc. We want to generate an image of what the product looks
    like in it's ideal state like in the Amazon-supplied images. 
    """
    completion = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": baseline},
            {"role": "user", "content": description},
            {"role": "user", "content": "DONE"},
        ]
    )
    print(completion.choices[0].message)
    print()
    
def extract_features_reviews(reviews):
    baseline = """
    I am a grad student in a Generative AI lab course. In this course, we have an 
    assignment to choose some products from Amazon, take their customer   
    reviews as textual data, and eventually create an accurate image of the product using
    generative image AI. I will give you all the original textual content from a 
    product. When I finish and type 'DONE', extract all the information 
    about what the product looks like so that I can create an accurate image later.
    Focus on visual information such as colors, texture, accessories, shape,
    decoration details, etc. We want to generate an image of what the product looks
    like in it's ideal state like in the Amazon-supplied images. 
    """
    completion = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": baseline},
            {"role": "user", "content": reviews},
            {"role": "user", "content": "DONE"},
        ]
    )
    print(completion.choices[0].message)
    print()
    
```
